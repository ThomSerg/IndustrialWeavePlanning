```python
import sys, os, json
```

```python
nr_item_types = [4,7,10]
nr_instances_per_problem = 10
```

```python
path_in = os.path.join(os.getcwd(), "..", "single_bin", "problems")
path_out = os.path.join(os.getcwd(), "problems")



for nr_item_type in nr_item_types:
    for y in range(100,900+1,50):
        for x in range(100,y+1,50):
            file_name = str(nr_item_type) + "_" + str(x) + "_" + str(y) 

            file = open(os.path.join(path_in, file_name + ".json"), "r")

            data = json.load(file)

            for data_instance_key, problem_instance in zip(data, range(nr_instances_per_problem)):

                data_instance = data[data_instance_key]

                machine_config = data_instance["machine_config"]
                machine_config["min_length"] = machine_config["max_length"]//2
                data_instance["machine_config"] = machine_config

                data[data_instance_key] = data_instance

                print(problem_instance, data_instance)


            json_object = json.dumps(data, indent=4)

            with open(os.path.join(path_out, file_name + ".json"), "w") as outfile:
                outfile.write(json_object)
```
