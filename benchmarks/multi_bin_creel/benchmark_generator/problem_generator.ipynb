# SLOPP + due-date + creel problem generator

```python
import sys, os, json
```

```python
nr_colorss = [1, 2, 3, 4, 5]
nr_item_types = [4,7,10]
nr_instances_per_problem = 10
max_nr_colors = 5
nr_deadlines = [1, 3, 6]

y = 400
x = 100
```

```python
path_in = os.path.join(os.getcwd(), "..", "..", "multi_bin", "problems", "_".join([str(i) for i in nr_deadlines]))
path_out = os.path.join(os.getcwd(), "..", "problems", "_".join([str(i) for i in nr_deadlines]))

for nr_colors in nr_colorss:

    for nr_item_type in nr_item_types:
        
        file_name = str(nr_item_type) + "_" + str(x) + "_" + str(y) 

        file = open(os.path.join(path_in, file_name + ".json"), "r")

        data = json.load(file)
        new_data = {}

        for data_instance_key in data:

            data_instance = data[data_instance_key]

            machine_config = data_instance["machine_config"]
            machine_config["max_creel_colors"] = nr_colors
            machine_config["max_creel_number"] = 5 # TODO
            machine_config["creel_switch_penalty"] = 10 # TODO
            data_instance["machine_config"] = machine_config

            data_instance["colors"] = [i%max_nr_colors for i in range(nr_item_type)]

            a = data_instance_key.strip().split("_")
            a.insert(-2, str(nr_colors))
            new_data["_".join(a)] = data_instance

            print(data_instance_key, data_instance)

        json_object = json.dumps(new_data, indent=4)

        with open(os.path.join(path_out, file_name + "_" + str(nr_colors) + ".json"), "w") as outfile:
            outfile.write(json_object)
```
